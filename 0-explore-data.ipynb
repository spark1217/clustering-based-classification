### Stats

##### Gender

```python
gender_counts = p_user['gender'].value_counts().sort_index()

plt.figure(figsize=(8, 6))
gender_counts.plot(kind='bar', color=['blue', 'pink'])
plt.xlabel('Gender')
plt.ylabel('Number of Users')
plt.title('Number of Users by Gender')
plt.xticks(ticks=[0, 1], labels=['Male', 'Female'], rotation=0)
plt.show()
```

```python
birth_counts = p_user['birth'].value_counts().sort_index()

plt.figure(figsize=(15, 6))
birth_counts.plot(kind='bar')
plt.xlabel('Birth year')
plt.ylabel('Number of Users')
plt.title('Number of Users by birth')
plt.xticks(rotation=90)
plt.show()
```

```python
season_counts = p_user['season_id'].value_counts().sort_index()

plt.figure(figsize=(10, 6))
season_counts.plot(kind='bar')
plt.xlabel('season')
plt.ylabel('Number of Users')
plt.title('Number of Users by season')
plt.xticks(rotation=90)
plt.show()
```

```python
# Count users by birth year and season_id
season_gender = p_user.groupby(['season_id', 'gender']).size().unstack(fill_value=0)

# Plot the data
season_gender.plot(kind='bar', figsize=(10, 8), colormap='viridis')
plt.xlabel('Season')
plt.ylabel('Number of Users')
plt.title('Number of Users by gender per Season')
plt.xticks(rotation=45)
plt.legend(title='Season ID')
plt.show()
```

```python
# Count users by birth year and season_id
grade = p_user.groupby(['season_id', 'grade_cd']).size().unstack(fill_value=0)
# grade = p_user['grade_cd'].value_counts().sort_index()
# Plot the data
grade.plot(kind='bar', figsize=(10, 8), colormap='viridis')
plt.xlabel('grade_cd')
plt.ylabel('Number of Users')
plt.title('Number of Users by grade_cd per Season')
plt.xticks(rotation=45)
plt.legend(title='grade_cd')
plt.show()
```

```python

season_price_sum = p_user[['bid','price','season_id']].groupby('season_id')['price'].sum()

# season_price_sum = p_user.groupby('season_id')['price'].sum()

# Plot the data
plt.figure(figsize=(12, 8))
season_price_sum.plot(kind='bar', color='skyblue')
plt.xlabel('Season ID')
plt.ylabel('Total Price')
plt.title('Total Price of Users by Season')
plt.xticks(rotation=0)
plt.show()
```
